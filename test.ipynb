# Handling Missing Values

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

# PERFORMING EDA 

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Count the number of apps per category
category_counts = df['prime_genre'].value_counts()

# Plot the top 10 categories
plt.figure(figsize=(10,6))
sns.barplot(x=category_counts.index[:10], y=category_counts.values[:10])
plt.title('Top 10 App Categories')
plt.xlabel('Category')
plt.ylabel('Number of Apps')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of app ratings
plt.figure(figsize=(10,6))
sns.histplot(df['user_rating'], bins=20, kde=True)
plt.title('Distribution of App Ratings')
plt.xlabel('User Rating')
plt.ylabel('Frequency')
plt.show()
```

# DRAWING INSIGHTS

```python
# Apps with the highest user ratings
top_rated_apps = df[df['user_rating'] == df['user_rating'].max()]
top_rated_apps[['track_name', 'user_rating', 'prime_genre']]
```

```python
# Correlation between app size and user rating
plt.figure(figsize=(10,6))
sns.scatterplot(x='size_bytes', y='user_rating', data=df)
plt.title('App Size vs. User Rating')
plt.xlabel('Size (Bytes)')
plt.ylabel('User Rating')
plt.show()
```
