```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

# PERFORMING EDA 

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Count the number of apps per category
category_counts = df['prime_genre'].value_counts()

# Plot the top 10 categories
plt.figure(figsize=(10,6))
sns.barplot(x=category_counts.index[:10], y=category_counts.values[:10])
plt.title('Top 10 App Categories')
plt.xlabel('Category')
plt.ylabel('Number of Apps')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of app ratings
plt.figure(figsize=(10,6))
sns.histplot(df['user_rating'], bins=20, kde=True)
plt.title('Distribution of App Ratings')
plt.xlabel('User Rating')
plt.ylabel('Frequency')
plt.show()
```

# DRAWING INSIGHTS

```python
# Apps with the highest user ratings
top_rated_apps = df[df['user_rating'] == df['user_rating'].max()]
top_rated_apps[['track_name', 'user_rating', 'prime_genre']]
```

```python
# Correlation between app size and user rating
plt.figure(figsize=(10,6))
sns.scatterplot(x='size_bytes', y='user_rating', data=df)
plt.title('App Size vs. User Rating')
plt.xlabel('Size (Bytes)')
plt.ylabel('User Rating')
plt.show()
```

```python
# Distribution of app prices
plt.figure(figsize=(10,6))
sns.histplot(df['price'], bins=20, kde=True)
plt.title('Distribution of App Prices')
plt.xlabel('Price ($)')
plt.ylabel('Number of Apps')
plt.show()
```

```python
# Count of free vs. paid apps
df['type'] = df['price'].apply(lambda x: 'Free' if x == 0 else 'Paid')
type_counts = df['type'].value_counts()

# Plotting
plt.figure(figsize=(6,6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Free vs. Paid Apps')
plt.show()

```

```python
# Average rating for free and paid apps
avg_rating = df.groupby('type')['user_rating'].mean()

# Plotting
avg_rating.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Average User Rating by App Type')
plt.xlabel('App Type')
plt.ylabel('Average User Rating')
plt.show()
```

# Conclusion 

Dominant Categories: Certain categories like Games and Entertainment have a higher number of apps.

User Ratings: The majority of apps have user ratings between 3.0 and 5.0.

App Size vs. Rating: There's no strong correlation between app size and user rating.

Free vs. Paid Apps: A significant majority of apps are free. However, paid apps tend to have slightly higher average user ratings.
